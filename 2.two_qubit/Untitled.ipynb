```python
from qiskit import (QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer)
import numpy as np
from qiskit.visualization import plot_histogram, plot_bloch_multivector


c = ClassicalRegister(2)
q = QuantumRegister(2)
qc = QuantumCircuit(q,c)
display(qc.draw(output = 'mpl'))
```

```python
c = ClassicalRegister(2)
q = QuantumRegister(2)
qc = QuantumCircuit(q,c)
```

```python
#Measurement Basis: AB
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.t(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Measurement Basis: Ab
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.tdg(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Measurement Basis: aB
qc.x(0)
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.t(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Measurement Basis: ab
qc.x(0)
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.tdg(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Bell pair + Measurement Basis (AB)
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.barrier()
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.t(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Bell pair + Measurement Basis(Ab)
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.barrier()
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.tdg(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```

```python
#Bell pair + Measurement Basis(aB)
qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.barrier()
qc.x(0)
qc.measure(0,0)
qc.s(1)
qc.h(1)
qc.t(1)
qc.h(1)
qc.measure(1,1)
display(qc.draw(output = 'mpl'))
```
